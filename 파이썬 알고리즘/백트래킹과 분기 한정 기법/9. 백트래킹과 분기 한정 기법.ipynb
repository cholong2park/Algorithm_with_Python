# 백트래킹이란?
- 이전에 공부한 **상태공간트리(state space tree)**는 어떤 문제를 해결하기 위한 모든 과정의 상태들을 각각 노드로 나타낸 트리를 말하는데, 백트래킹은 이러한 상태공간트리를 보다 효율적으로 탐색하는 방법을 제공한다. 백트래킹은 다음과 같은 전략을 사용한다.

- 백트래킹의 기본 골격은 **깊이 우선 탐색(DFS)**인데, 이것은 트리의 모든 노드를 체계적으로 방문할 수 있는 완전 탐색 전략의 하나이다.
- 백트래킹은 상태공간트리를 루트에서부터 깊이 우선으로 탐색해 내려오다가 단말 노드까지 가기 전에 방문한 어떤 노드(후보해)가 문제에서 요구하는 해가 아니라고 판단되면 더 이상 탐색하지 않고 이전 노드로 되돌아 나오는(backtracking) 전략을 사용한다.

이를 통해, 완전 탐색과는 달리 전체 상태공간트리의 많은 노드들을 탐색하지 않고 미리 **가지치기(pruning)**할 수 있고, 따라서 탐색의 효율을 높일 수 있다. 주어진 문제를 백트래킹으로 해결하는 알고리즘의 설계과정은 다음과 같다.

1. 해를 탐색하는 과정은 루트노드에서 시작한다. 루트 노드도 하나의 후보해이다.
2. 현재의 후보해에서 선택이 가능한 다음 후보해의 리스트를 구한다.
3. 후보해 리스트의 각 해에 대해 다음을 수행한다.
4. 현재의 해가 문제에서 요구하는 조건을 만족시키면 2~3을 계속 수행한다. 이것은 현재 노드의 서브 트리를 계속 탐색하는 것이다. 만약 조건이 만족되지 않으면 이전 후보해로 되돌아가(backtracking) 다음 후보해를 처리한다.
5. 2~4 과정을 최종해를 구할 때까지 반복한다. 만약 모든 가능한 해를 탐색해도 해가 없다면 이 문제는 해가 없는 것이므로 종료한다.

백트래킹에서 다루는 문제들은 보통 여러 개의 해가 존재할 수 있다. 문제에 따라서는 하나의 해만 찾으면 될 수도 있고, 모든 해를 구해 이들 중에서 **최적해(optional solution)**를 찾도록 요구할 수도 있다. 이렇게 최적해를 요구하는 문제를 최적화 문제라고 한다. 예를 들어, 미로탐색도 단순히 하나의 탈출 경로를 찾는 문제가 될 수도 있고, 모든 탈출 경로를 모두 찾아 가장 짧은 것을 구하는 최적화 문제가 될 수도 있다.

# 분기 한정 기법이란?
- 이러한 백트래킹 기법은 상태공간트리의 많은 노드들을 가지치기하여 효율을 높일 수는 있지만, 문제에 따라 여전히 비효율적인 경우가 많다. 예를 들어, 최적화 문제에는 최적해가 상태공간트리의 어디에 있는지 알 수 없으므로 대부분의 노드를 탐색해야 하고, 따라서 백트래킹이 크게 효과적이지 않을 수 있다. **분기 한정(branch and bound)** 기법은 이러한 백트래킹의 문제를 보완하고자 하는 방법이다.
